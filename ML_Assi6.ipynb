```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
df = pd.read_csv('data_Asig6.csv')
#1st plot
df.plot(kind="scatter",x='MinTemp',y='MaxTemp')

```

```python
print("*2"+"*"*50+"Kernel Density Estimation"+"*"*50)
sns.jointplot(x='MinTemp', y='Evaporation', data=df, kind='kde')
```

```python
print("*3"+"*"*50+"Kernel Density Estimation"+"*"*50)
sns.distplot(df['Evaporation'],bins=15)
```

```python
print("*4"+"*"*50+"Histogram"+"*"*50)
sns.distplot(df['Evaporation'],bins=15, kde=False, rug=True)
```

```python
print("*5"+"*"*50+"*"*50)
sns.kdeplot(df['Evaporation'], shade=True)
```

```python
print("*6"+"*"*50+"*"*50)
sns.kdeplot(df['Evaporation'], label='original')
sns.kdeplot(df['Evaporation'], bw=1, label='bw=1')
sns.kdeplot(df['Evaporation'], bw=2, label='bw=2')
plt.legend()
```

```python
print("*7"+"*"*100)
x=df['Evaporation'].T
y=df['MinTemp'].T
with sns.axes_style('white'):
    sns.jointplot(x = x, y = y, kind = 'hex', color = 'k')
```

```python
print("*8"+"*"*50+"Stripplot"+"*"*50)
sns.stripplot(x = 'Evaporation', y='Sunshine', data=df)
```

```python
print("*9"+"*"*50+"Swarmplot"+"*"*50)
sns.swarmplot(x = 'MinTemp', y='MaxTemp', data=df)
```

```python
print("*10"+"*"*50+"Boxplot"+"*"*50)
sns.boxplot(x = 'RainTomorrow', y='Rainfall', data=df)
```

```python
print("*11"+"*"*50+"Stripplot version"+"*"*50)
sns.stripplot(x = 'RainTomorrow', y='Rainfall', data=df,jitter=True)
```

```python
print("*11"+"*"*50+"Stripplot version"+"*"*50)
sns.stripplot(x = 'WindGustDir', y='WindGustSpeed', data=df,jitter=True)
```

```python
print("*12"+"*"*50+"Barplot"+"*"*50) 
sns.barplot(x = 'WindGustDir', y='WindGustSpeed', data=df)
```

```python
print("*13"+"*"*50+"Countplot"+"*"*50) 
sns.countplot(x = 'WindGustDir', data=df,palette='Greens_d')
```
